# **Exercise 2**

First we import the required libararies

```python
import matplotlib.pyplot as plt
import numpy as np
```

We create an array with the range of x, the data starts at -6, ends at -6, there is a difference of 0.5 between each data point. This prodices an array with 24 elements. We then Seperate the components of the required function: **f(x) = sin(x) + cos(2x)/2 + sin(3x)/3 + cos(4x)/4**.

```python
x = np.arange(-6, 6, 0.1)
fx1 = np.sin(x) #assigning the sinx function to a variable
fx2 = np.cos(2*x)/2 #assigning cos2x/2 function to a variable
fx3 = np.sin(3*x)/3 #assigning sin3x/3 function to a variable
fx4 = np.cos(4*x)/4 #assigning cos4x/4 function to a variable
```

We then take the sum of all the individual functions to produce the funtion **f(x)**.

```python
fx = fx1 + fx2 + fx3 + fx4
```

The first graph shows the function **f(x)**. The markers used to plot the data points are **circles**, the line of the plot is **dashed**, and the color of the plot is **red**.

```python
plt.plot(x, fx, marker = 'o', linestyle = '--', color = 'red')
plt.xlim([-6, 6])
plt.xlabel('X')
plt.ylabel('Y')
```

The second graph shows the function **f(x)**. The markers used to plot the data points are **triangles**, the line of the plot is **dotted**, and the color of the plot is **gray**.

```python
plt.plot(x, fx, marker = '^', linestyle = ':', color = 'gray')
plt.xlabel('X')
plt.ylabel('Y')
```

The third graph shows the function **f(x)**. The markers used to plot the data points are **arrows**, the line of the plot is **solid**, and the color of the plot is **magenta**.

```python
plt.plot(x, fx, marker = '1', linestyle = '-', color = 'magenta')
plt.xlabel('X')
plt.ylabel('Y')
```

While all three plots yield the same graphical results, the differences in each lies in the display of each graph. The first graph is displayed as a **red**, **dashed** function with **circles** representing the data points. The second graph is a **dotted**, **gray** function, and the data points are represented with **triangles**. The final graph is a **solid** **magenta**, and the data points appear as **arrows**. Although all plots are similar, manipulation of the display parameters yields slightly different results.

# **Exercise 3**

Create the arrays with the data to be plotted

```python
# percentage of male and female passengers
gender = [55, 45]
#Different age intervals of the passengers
age_labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-99', '90-99']
#The number of people in each age interval
age = [16, 23, 347, 428, 334, 398, 923, 1015, 216, 11]
```

We then create a chart that has both a bar graph and a plot. We make the plot's line dashed, and its markers circular. The axis titles are set to bold and the x-axis labels are rotated 45 degrees. 

```python
plt.bar(age_labels, age)
plt.plot(age_labels, age, color = 'red', marker = 'o', linestyle = '--')
plt.xlabel('Age', fontweight = 'bold')
plt.ylabel('Quantity', fontweight = 'bold')
plt.xticks(rotation = 45)
plt.show
```

We then create the labels for the pie chart, set a title, and display the proportion of each section.

```python
gender_labels = ['Male', 'Female']
plt.title('Gender', fontweight = 'bold')
exp = [0.06, 0]
plt.pie(gender, labels = gender_labels, shadow = True, explode = exp, autopct = '%i%%')
plt.show()
```

After assigning the labels and frequencies into arrays, the first produced graph is a combination of two graphical methods. A bar chart is used to represent the frequency of each age range, or how many people lie in each range. A plot is then added and shows the frequency as a scatter plot. This type of plot is often called a pareto chart and represents the frequency of occurence of each designated range. After plotting the frequency, a pi plot is used to show the distinction of male to female participants which are 55% and 45% respectively as shown by the pie chart above. Both graphs are related to the same group of people while each of them reresent a unique set of information.

```python
counter = 0
arr = []

for i in range(len(age)):
    for j in range(age[counter]):
        arr.append(counter)
    counter = counter + 1
```

We create an expanded array that containes each occurence of each index, that is for the first age range (index 0) we repeat index 0 16 times before beginning to insert occurences of the next index

```python
mean = np.mean(arr)
print(mean)
```

```python
std = np.std(arr)
print(std)
```

the above mean shows that the mean lies in the 5th index, or age range 50 - 59 years old

```python
import scipy.stats as stats

x_axis = np.arange(0, 10, 1)
plt.plot(x_axis, stats.norm.pdf(x_axis, mean, std))
plt.show()
```

The normal distribution above indicates that the majority of people are between indices **4 and 7**, that is the majority of people are between the ages of **40 and 79**. The mean being **5.33** indicated that the mean lies in age range **50 - 59**. The standard deviation of **1.84** indicates that the data is relatively clustered around the mean. A larger standard deviation would have produced a flatter, more distributed curve where data is less clustered around the mean of **5.33**.
